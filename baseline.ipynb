### Baseline (popularity item)

```python
# =====================================================================
# 1. TẠO BASELINE: TÌM 10 MÓN BÁN CHẠY NHẤT TRONG TẬP TRAIN
# =====================================================================
print("1. Đang tính toán Baseline (Top 10 sản phẩm Quốc dân)...")

# Chỉ đếm những sự kiện 'purchase'
top_10_popular = (
    train_set
    .filter(pl.col('event_type') == 'purchase')
    .group_by('item_id')
    .agg(pl.len().alias('purchase_count'))
    .sort('purchase_count', descending=True)
    .head(10)
    .collect(streaming=True) # Ép LazyFrame thực thi
    .get_column('item_id')
    .to_list()
)

print(f"-> Đã tìm ra Top 10: {top_10_popular}")

# =====================================================================
# 2. HÀM GỢI Ý BASELINE (Cực kỳ đơn giản)
# =====================================================================
def recommend_baseline(target_user_id):
    # Hệ thống ngốc nghếch: Bất kể khách hàng là ai, cứ nhét 10 món này cho họ
    return top_10_popular

# =====================================================================
# 3. ĐÁNH GIÁ TRÊN TẬP VALIDATION
# =====================================================================
print("\n2. Đang đánh giá Precision@10 trên tập Validation...")

# Rút tập Validation ra thành DataFrame để duyệt
ground_truth = (
    val_set.filter(pl.col('event_type') == 'purchase')
    .group_by('customer_id')
    .agg(pl.col('item_id').alias('actual_items'))
    .collect(streaming=True)
)

hits_total = 0.0
users_eval = 0

# Chạy đánh giá
for row in ground_truth.iter_rows(named=True):
    user_id = row['customer_id']
    actual = set(row['actual_items'])
    
    # Dự đoán bằng Baseline
    preds = set(recommend_baseline(user_id))
    
    # Tính điểm
    hits = len(preds.intersection(actual))
    hits_total += hits / 10.0
    users_eval += 1

baseline_precision = (hits_total / users_eval) * 100 if users_eval > 0 else 0
print(f"\n[KẾT QUẢ BASELINE] Mean Precision@10: {baseline_precision:.2f}%")
print(f"Đã đánh giá trên tổng cộng {users_eval} người dùng.")
```

### Upgrade 1 (Using quantity and collaborative filtering)

```python
import polars as pl

print("--- THỐNG KÊ SỐ LIỆU THÔ: PHÂN TÍCH BIẾN QUANTITY ---")

# Tính toán các chỉ số thống kê theo từng loại hành vi
quantity_eda = (
    train_set
    .group_by('event_type')
    .agg([
        # 1. Tổng số bản ghi của hành vi đó
        pl.len().alias('total_interactions'),
        
        # 2. Số lượng bị khuyết (Null)
        pl.col('quantity').null_count().alias('null_count'),
        
        # 3. Giá trị trung bình và trung vị
        pl.col('quantity').mean().round(2).alias('mean_quantity'),
        pl.col('quantity').median().alias('median_quantity'),
        
        # 4. Giá trị cực đại (Để săn lùng Outliers)
        pl.col('quantity').max().alias('max_quantity'),
        
        # 5. Đếm xem có bao nhiêu dòng khách tương tác với số lượng lớn hơn 1
        (pl.col('quantity') > 1).sum().alias('rows_greater_than_1')
    ])
    .with_columns(
        # Tính tỷ lệ % xuất hiện số lượng lớn
        (pl.col('rows_greater_than_1') / pl.col('total_interactions') * 100).round(2).alias('pct_greater_than_1_%')
    )
    .collect(engine='streaming')
)

print(quantity_eda)
```

```python

print("1. Đang xây dựng Ma trận Đồng xuất hiện (Co-occurrence) và Phổ biến...")

# 1. Tính độ phổ biến toàn cục (Để fallback và Cold-start)
global_popularity = (
    train_set.filter(pl.col('event_type') == 'purchase')
    .group_by('item_id')
    .agg(pl.col('quantity').fill_null(1).sum().alias('total_sold'))
    .sort('total_sold', descending=True)
    .with_columns((pl.col('total_sold') / pl.col('total_sold').max()).alias('pop_score'))
    .collect(streaming=True)
)
cold_start_items = global_popularity.head(10).get_column('item_id').to_list()

# 2. Xây dựng Ma trận Đồng xuất hiện (Market Basket Analysis)
purchases = (
    train_set.filter(pl.col('event_type') == 'purchase')
    .select(['bill_id', 'item_id'])
    .drop_nulls()
)

# Đếm mẫu số: Item A xuất hiện trong bao nhiêu hóa đơn?
item_bill_counts = purchases.group_by('item_id').agg(pl.len().alias('bill_count_A'))

# Self-join để tìm các cặp (A, B) trong cùng hóa đơn
co_occurrence_matrix = (
    purchases.rename({"item_id": "item_A"})
    .join(purchases.rename({"item_id": "item_B"}), on='bill_id')
    .filter(pl.col('item_A') != pl.col('item_B'))
    # Đếm số hóa đơn chứa cả A và B
    .group_by(['item_A', 'item_B'])
    .agg(pl.len().alias('cooc_count'))
    # Ghép với mẫu số để tính Confidence
    .join(item_bill_counts, left_on='item_A', right_on='item_id')
    .with_columns(
        (pl.col('cooc_count') / pl.col('bill_count_A')).alias('confidence_score')
    )
    # Lọc giữ lại Top 20 mặt hàng mua kèm tốt nhất cho mỗi item_A để tiết kiệm RAM
    .sort(['item_A', 'confidence_score'], descending=[False, True])
    .group_by('item_A')
    .head(20)
    .collect(streaming=True)
)
print("-> Hoàn tất Ma trận Offline!")
```

```python
BETA = 0.8  # Trọng số cho độ liên quan cá nhân (Mua kèm)
GAMMA = 0.2 # Trọng số cho xu hướng đám đông (Phổ biến)
TOP_K = 10
```

```python
import concurrent.futures
from tqdm.auto import tqdm # Thư viện tạo thanh tiến trình (Progress Bar)
import polars as pl

# Đảm bảo pool_pop đã được tính 1 lần duy nhất ở ngoài (để không phải tính lại 800k lần)
pool_pop_global = global_popularity.head(100).select([
    pl.col('item_id').alias('candidate_item'),
    pl.col('pop_score')
]).cast({'candidate_item': pl.Utf8})

print("1. Đang lọc Ground Truth (Đán án) từ tập Validation...")
ground_truth = (
    val_set.filter(pl.col('event_type') == 'purchase')
    .group_by('customer_id')
    .agg(pl.col('item_id').alias('actual_items'))
    .collect(streaming=True)
)

val_users_list = ground_truth.get_column('customer_id').to_list()
total_users = len(val_users_list)
print(f"-> Tổng số User cần đánh giá: {total_users}")


print("\n2. Đang quét tập Train để tính sẵn lịch sử cho toàn bộ Val Users...")

all_val_history_df = (
    train_set.filter(pl.col('customer_id').is_in(val_users_list))
    
    # BƯỚC 1: Gom các dòng trùng khít GIỜ-PHÚT-GIÂY do lỗi hệ thống ghi trùng
    # Nếu cùng 1 thời điểm có nhiều log qty=1, ta cộng tổng lại thành số lượng thực tế tại thời điểm đó
    .group_by(['customer_id', 'item_id', 'event_time', 'event_type'])
    .agg(pl.col('quantity').fill_null(1).sum().alias('exact_moment_qty'))
    
    # BƯỚC 2: Tính điểm cho từng dòng thời gian (Áp dụng Clip chặn ngưỡng cho từng lượt tương tác)
    .with_columns([
        pl.when(pl.col('event_type').is_in(['purchase', 'add_to_cart']))
        .then(pl.col('exact_moment_qty').clip(upper_bound=5))
        .otherwise(pl.col('exact_moment_qty').clip(upper_bound=2))
        .alias('capped_qty'),
        
        pl.when(pl.col('event_type') == 'purchase').then(3.0)
        .when(pl.col('event_type') == 'add_to_cart').then(1.5)
        .otherwise(0.5).alias('event_weight')
    ])
    .with_columns((pl.col('capped_qty') * pl.col('event_weight')).alias('interaction_score'))
    
    # BƯỚC 3: Cộng tổng xuyên suốt lịch sử để đo "độ cuồng" tổng thể của khách
    .group_by(['customer_id', 'item_id'])
    .agg(pl.col('interaction_score').sum().alias('user_preference'))
    .collect(engine='streaming')
)

# Rút riêng danh sách các mặt hàng đã Add to Cart
all_atc_df = (
    train_set.filter((pl.col('customer_id').is_in(val_users_list)) & (pl.col('event_type') == 'add2cart'))
    .select(['customer_id', 'item_id'])
    .unique()
    .collect(streaming=True)
)


print("\n3. Đang nạp dữ liệu vào RAM (Hash Map) để tra cứu siêu tốc...")

# Hàm chuẩn hóa key cho các phiên bản Polars khác nhau
def parse_dict(partitioned_data):
    return {k[0] if isinstance(k, tuple) else k: df for k, df in partitioned_data.items()}

history_dict = parse_dict(all_val_history_df.partition_by('customer_id', as_dict=True))
atc_dict = parse_dict(all_atc_df.partition_by('customer_id', as_dict=True))

# Chuyển ATC items thành list luôn cho nhẹ
atc_list_dict = {k: v.get_column('item_id').to_list() for k, v in atc_dict.items()}

# Dọn dẹp rác bộ nhớ
del all_val_history_df, all_atc_df
import gc; gc.collect()


def fast_generate(target_user_id):
    # Lấy dữ liệu từ RAM thay vì chui vào train_set lọc
    if target_user_id not in history_dict:
        return cold_start_items[:TOP_K]
        
    user_history_df = history_dict[target_user_id]
    atc_items = atc_list_dict.get(target_user_id, [])
    
    # Tính Co-occurrence (Chỉ sử dụng df nhỏ trong RAM)
    pool_cooc = (
        user_history_df
        .join(co_occurrence_matrix, left_on='item_id', right_on='item_A', how='inner')
        .with_columns((pl.col('user_preference') * pl.col('confidence_score')).alias('weighted_score'))
        .group_by('item_B')
        .agg(pl.col('weighted_score').sum().alias('user_cooc_score'))
        .rename({'item_B': 'candidate_item'})
    )
    
    if pool_cooc.height > 0:
        max_score = pool_cooc.select(pl.col('user_cooc_score').max()).item()
        pool_cooc = pool_cooc.with_columns((pl.col('user_cooc_score') / max_score).alias('user_cooc_score_norm'))
    else:
        pool_cooc = pool_cooc.with_columns(pl.col('user_cooc_score').alias('user_cooc_score_norm'))

    pool_cooc = pool_cooc.cast({'candidate_item': pl.Utf8})

    final_ranking = (
        pool_cooc
        .join(pool_pop_global, on='candidate_item', how='outer_coalesce')
        .fill_null(0.0)
        .with_columns((pl.col('user_cooc_score_norm') * BETA + pl.col('pop_score') * GAMMA).alias('final_score'))
        .with_columns(
            pl.when(pl.col('candidate_item').is_in(atc_items))
            .then(pl.col('final_score') + 2.0)
            .otherwise(pl.col('final_score')).alias('final_score')
        )
        # Chốt cứng tiêu chí phụ (candidate_item) để tránh nhảy điểm
        .sort(['final_score', 'candidate_item'], descending=[True, False]) 
    )

    return final_ranking.head(TOP_K).get_column('candidate_item').to_list()


# print("\n4. Đang khởi chạy hệ thống Đa luồng đánh giá (Multi-threading)...")

# def evaluate_single_user(row):
#     user_id = row['customer_id']
#     actual = set(row['actual_items'])
    
#     preds = set(fast_generate(user_id))
#     hits = len(preds.intersection(actual))
#     return hits / 10.0

# # Chuyển đổi ground_truth thành list of dicts để đưa vào luồng dễ dàng
# rows_to_eval = ground_truth.to_dicts()

# hits_total = 0.0

# with concurrent.futures.ThreadPoolExecutor() as executor:
#     # Lệnh map sẽ tự phân bổ công việc ra các core của CPU
#     results = list(tqdm(executor.map(evaluate_single_user, rows_to_eval), total=total_users, desc="Đang đánh giá Precision@10"))

# hits_total = sum(results)
# precision_score = (hits_total / total_users) * 100

# print("-" * 50)
# print(f"[KẾT QUẢ CUỐI CÙNG] Mean Precision@10: {precision_score:.2f}%")
# print("-" * 50)
```

```python
import pickle
import concurrent.futures
from tqdm.auto import tqdm
import polars as pl

# Đảm bảo sửa lỗi chữ 'add2cart' thành 'add_to_cart' ở bước lấy dữ liệu ATC trước đó của bạn:
all_atc_df = (
    train_set.filter((pl.col('customer_id').is_in(val_users_list)) & (pl.col('event_type') == 'add_to_cart'))
    .select(['customer_id', 'item_id'])
    .unique()
    .collect(streaming=True)
)

# =====================================================================
# BƯỚC ĐỘT PHÁ 4: CHẠY ĐA LUỒNG GENERATE DỰ ĐOÁN (KHÔNG CHẤM ĐIỂM)
# =====================================================================
print("\n4. Đang khởi chạy hệ thống Đa luồng để tạo dự đoán cho từng User...")

def predict_single_user(user_id):
    """Hàm chạy cho từng luồng: Trả về cặp (user_id, danh sách_dự_đoán)"""
    preds = fast_generate(user_id)
    return user_id, preds

# Khởi tạo Dictionary để chứa kết quả cuối cùng
final_predictions = {}

# Sử dụng ThreadPoolExecutor để tận dụng tối đa CPU
with concurrent.futures.ThreadPoolExecutor() as executor:
    # Chạy song song hàm predict_single_user cho toàn bộ danh sách khách hàng
    results = list(tqdm(
        executor.map(predict_single_user, val_users_list), 
        total=total_users, 
        desc="Đang xử lý xuất dữ liệu dự đoán"
    ))

# Chuyển đổi list of tuples thành Dictionary dạng {customer_id: [items]}
final_predictions = dict(results)

print(f"-> Đã tạo xong dự đoán cho {len(final_predictions):,} người dùng.")

# =====================================================================
# BƯỚC 5: XUẤT DỮ LIỆU RA FILE PICKLE
# =====================================================================
pickle_filename = "user_predictions_top10.pkl"
print(f"\n5. Đang đóng gói và ghi dữ liệu ra file: {pickle_filename}...")

with open(pickle_filename, 'wb') as f:
    # Ghi đè toàn bộ dictionary vào file nhị phân pickle với giao thức tối ưu nhất
    pickle.dump(final_predictions, f, protocol=pickle.HIGHEST_PROTOCOL)

print(f"==================================================")
print(f" THÀNH CÔNG: Đã lưu file dự đoán tại '{pickle_filename}'")
print(f"==================================================")
```
